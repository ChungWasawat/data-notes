### Set

````python
set1= {"ant", "bird", "cat", "bird"}
set2 = set(["bird", "cat", "dog", "cat"])

set1.add(20)
set1.remove(20)
print(set1)

if set1 & set2 == set1.intersection(set2):
    print("intersected")

if set1 | set2 == set1.union(set2):
    print("unioned")

print(set1.difference(set2), set2.difference(set1))
print(set1.symmetric_difference(set2), set2.symmetric_difference(set1))
````

### \*args and **kwargs

````python
# learn more about positional parameter (*, /)
# *args and **kwargs allow you to pass multiple arguments or keyword arguments to a function.
# **kwds is optional choice for the function
def test_func(a, b, *c):
    print(list(c))
test_func(1, 2, 3, 4 ,5)
def test_func(a, b, **option):
    if option.get("c") == 0:
        print(a+a)
    elif option.get('c') == 1:
        print(b+b)
    else:
        print(a+b)
test_func('op', 'tion')
test_func('op', 'tion', c=1)

def test_func(a, *b):
    print(type(b))  # it's tuple
test_func(1,2,3)
def test_func(a, **b):
    print(type(b), b.keys(), b.values()) # it's dict
test_func(a=1,b=2)
````

### Decorator

````python
def repeater(old_function):                 #function as parameter here
    # See learnpython.org/en/Multiple%20Function%20Arguments for how *args and **kwds works
    # not care about argument of old function as using *args, **kwargs
    def new_function(*args, **kwds):
        old_function(*args, **kwds)
        old_function(*args, **kwds)
    return new_function
@repeater
def multiply(num1, num2):
    print(num1 * num2)
multiply(5, 3)


def multiply(multiplier):                   #variable from decorator's parameter
    def multiply_generator(old_function):   #function as parameter here
        def new_function(*args, **kwds):
            return multiplier * old_function(*args, **kwds)
        return new_function
    return multiply_generator # it returns the new generator

@multiply(4) # multiply is not a generator, but multiply(3) is
def return_num(num): #as old function
    return num-1
# Now return_num is decorated and reassigned into itself
return_num(6) # should return 15
````

````python
class Dog:
  """details for dog"""
  def __init__(self, colour="white"):
      self.colour = colour

  def showAttribute(self):
      return self.colour

  def bark(self):
      print("woof")

  def __repr__(self) -> str:
      return f'MyClass("{self.colour}")'
# inheritance
class Puddle(Dog):
  """subclass of dog"""
  def __init__(self, super, fur="square"):
      self.fur = fur

  def bark():
      print("box")

boo = Dog()

print(hasattr(Dog, 'colour'))               #check if an object has a specific attribute
print(id(boo))                              #returns the unique identifier for an object
print(type(boo))                            #return data type of object
print(Dog())                                #returns a string representation of an object
print(callable(boo.bark()))                 #check if an object appears callable
print(issubclass(Puddle, Dog))              #check if a class is a subclass of another class
print(isinstance(Dog(), Dog))               #check if an object is an instance of a specified class or a tuple of classes
print(Dog.__doc__)                          #access the documentation string (docstring) of an object
print(__name__)
#When a Python file is run directly, the __name__ variable is set to "__main__".
#When a Python file is imported as a module, the __name__ variable is set to the name of the module.
````

## Interesting Packages


#### dataclass

````python
# for pros and cons, can see more here: https://www.startdataengineering.com/post/code-patterns/#python-helpers
from dataclasses import asdict, dataclass
@dataclass
class RedditPostData:
    """Dataclass to hold reddit post data.
    Args:
        title (str): Title of the reddit post.
        score (int): Score of the reddit post.
        url (str): URL of the reddit post.
        comms_num (int): Number of comments on the reddit post.
        created (str): Datetime (string repr) of when the reddit
             post was created.
    """
    title: str
    score: int
    url: str
    comms_num: int
    created: str
    text: str
@dataclass
class TwitterTweetData:
    """Dataclass to hold twitter post data.
    Args:
        text (str): Text of the twitter post.
    """
    text: str
@dataclass
class SocialMediaData:
    """Dataclass to hold social media data.
    Args:
        id (str): ID of the social media post.
        text (str): Text of the social media post.
    """
    id: str
    source: str
    social_data: RedditPostData | TwitterTweetData # social_data can be one of the Reddit or Twitter data types
a = TwitterTweetData(text="Hello guys")
b = SocialMediaData(id="abc123", source="yagoogoo", social_data=a)
print(b.social_data)
````

#### contextmanagaer

````python
# contextmanagaer
from contextlib import contextmanager
@contextmanager
def open_file(file, mode):
    f = open(file, mode)
    try:
        yield f
    finally:
        # if it's database connection, can put commit and close here
        f.close()
with open_file('test.txt', 'r') as f:
    # if it's database connection, 'execute' can be put here
    a = f.readlines()
    print(a)
````
